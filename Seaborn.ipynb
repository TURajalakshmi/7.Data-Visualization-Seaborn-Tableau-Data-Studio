```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
dataset = pd.read_csv("Placement.csv")
dataset
```

```python
print("The number of rows:", dataset.shape[0])
print("The number of column:", dataset.shape[1])
```

### Univarite analysis

```python
sb.distplot(dataset["hsc_p"],hist=True,kde=True,kde_kws={'color':'purple'})
```

### Bivarite analysis: Distribution plot

```python
sb.jointplot(x = 'hsc_p', y = 'salary', data = dataset)
plt.show()
```

```python
sb.jointplot(x = 'hsc_p', y = 'salary', data = dataset, kind ='hex')
plt.show()
```

```python
sb.jointplot(x = 'hsc_p', y = 'salary', data = dataset,kind='kde')
plt.show()
```

```python
sb.pairplot(dataset,hue='gender', diag_kind='kde', kind='scatter', palette = 'husl')
plt.show()
```

### For categorical plot

```python
sb.stripplot(x = 'hsc_b', y = 'hsc_p', data = dataset)
plt.show()
```

```python
sb.stripplot(x = 'gender', y = 'hsc_p', data = dataset, jitter=True)
plt.show()
```

```python
sb.swarmplot(x = 'gender', y = 'hsc_p', data = dataset)
plt.show()
```

```python
sb.violinplot(x = 'gender', y = 'hsc_p', data=dataset, hue='gender')
plt.show()
```

```python
sb.violinplot(x = 'degree_t', y = 'salary', data=dataset, hue='gender')
plt.show()
```

```python
sb.boxplot(x='hsc_b', y='hsc_p', data=dataset, hue='gender')
plt.show()
```

```python
sb.barplot(x='hsc_b', y='hsc_p', data=dataset, hue='gender')
plt.show()
```

```python
sb.countplot(x='degree_t', data=dataset, hue='gender')
plt.show()
```

```python
sb.pointplot(x='hsc_b', y='hsc_p', data=dataset, hue='gender')
plt.show()
```

```python
sb.boxplot(data=dataset, orient='h')
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('exercise')
sb.catplot(x='time', y='pulse', hue='kind', kind='violin', col='diet', data = df);
plt.show()
```

```python
df=sb.load_dataset('tips')
sb.regplot(x='total_bill', y='tip', data=df)
sb.lmplot(x='total_bill', y='tip', data=df)
plt.show()
```

```python
df = sb.load_dataset('anscombe')
sb.lmplot(x='x', y='y', data=df.query("dataset=='II'"))
plt.show
```

```python
df = sb.load_dataset('anscombe')
sb.lmplot(x='x', y='y', data=df.query("dataset=='II'"),order=2)
plt.show
```

```python
df = sb.load_dataset('tips')
g = sb.FacetGrid(df,col='time')
g.map(plt.hist, "tip")
plt.show()
```

```python
df = sb.load_dataset('iris')
g = sb.PairGrid(df)
g.map(plt.scatter);
plt.show()
```
